```python
corr = d.corr()
sns.heatmap(corr,annot=True,cmap='Reds')
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='State',data=df,palette='rocket_r',order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
# Conclusion :
# The above Graph shows that , count of Customers.
# here : The customers from California state are high in number and from Wyoming are low in number . 
# here : From CALIFORNIA State the supermatket gains more profit and from WYOMING gains lower profit .
# By visualizing this the supermarket can focus to improve the market products and Quality more in
# the states where more Customers are buying products .
```

```python
plt.figure(figsize=(12,10))
d['Region'].value_counts().plot.pie(autopct='%.2f')
plt.show()
```

```python
df.groupby(['Region']).agg({'Sales':'sum'}).plot(kind='bar',legend=True,color='Green')
```
